 ### TensorFlow Mechanics
 1. Build graph using TensorFlow operations
 2. feed data and run graph : sess.run
 3. update variables in the graph

#### (1) 그래프 구현

```python
import tensorflow as tf
tf.__version__
```

```python
x_train = [1,2,3]
y_train = [1,2,3]

#tensorflow가 학습하는 과정에서 자기가 변경시킨다.
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x_train* W + b
```

```python
#cost/loss function : square 제곱
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost) #cost를 minimize해라
```

#### (2) session 만들기

```python
sess = tf.Session()

#W,b 라는 tensorflow 변수를 사용하기 전에 실행해야 할것
sess.run(tf.global_variables_initializer())
```

```python
#Fit the line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
#Placeholders

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train],
            feed_dict ={X:[1,2,3], Y:[1,2,3]})
    if step%20 ==0:
        print(step, cost_val, W_val, b_val)
```
